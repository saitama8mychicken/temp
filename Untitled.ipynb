```python
file = " Let this be first.this is a normal file. It has somne of tghe upper words like engine, interior. The other words follo are as foll and has a performance of good mileage price. For the comfort of interior price of the engine for a good performance"
```

```python
def file_analyser(file, keyword_list = ['engine', 'mileage', 'comfort', 'performance', 'interior', 'maintenance', 'price']):
    import matplotlib.pyplot as plt
    import nltk
    from nltk.tokenize import word_tokenize
    from nltk.tokenize import sent_tokenize
    keyword_filtered = []
    for item in sent_tokenize(file):
        tokens = word_tokenize(item)
        # to filter out the sentences
        for word in tokens:
            if word in keyword_list:
                keyword_filtered.append(item)
                break
            else:
                continue
    print('################## The sentence with keywords provided are #################')
    print()
    for item in (keyword_filtered):
        print(item)
    print()
    print('  '*10+'$'*30)
    print()
    # creating dictionary to store frequencies of words for bar plot
    new_li = dict()
    for sent in sent_tokenize(file):
        tokens = word_tokenize(sent)
        for word in tokens:
            if word in keyword_list:
                if word in new_li:
                    new_li[word] += 1
                else:
                    new_li[word] = 1
    fig1, ax1 = plt.subplots()
    ax1.bar(new_li.keys(), new_li.values())
    ax1.set_title('comparing frequencies')
    #plotting the same over wordcloud
    from wordcloud import WordCloud
    fig2, ax2 = plt.subplots()
    ax2.set_title('Wordcloud')
    wordcloud = WordCloud(background_color = 'black', random_state = 1).generate_from_frequencies(new_li)
    ax2.axis('off')
    ax2.imshow(wordcloud)
    
    
    
```

```python
file_analyser(file, keyword_list = [ 'engine', 'mileage', 'comfort', 'performance', 'interior', 'maintenance', 'price'])
```
